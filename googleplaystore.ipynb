Finding Average Rating


```python
Avg_rating=float(str(df['Rating'].sum()/len(df['Rating']))[:4])
Avg_rating
```

How many Apps with rating 5 are there in the Dataset

```python
R5=0
for i in df['Rating']:
  if(i==5.0):
    R5+=1
print("There are ", R5 ," Applications with rating 5.")
```

How many Apps are with rating between 4.0 and 4.5

```python
c=0
for i in df['Rating']:
  if(int(i)>=4.0 or int(i)<=4.5):
    c+=1
print("There are ", c ," Applications with rating between 4.0 and 4.5.")
```

Total no.of applications in Art and Design category

```python
c=0
for i in df['Category']:
  if i=='ART_AND_DESIGN':
    c=c+1
print('There are ',c, 'Applications in teh catagory ART_AND_DESIGN')
```

```python
f=0
p=0
for i in df['Type']:
  if i=='Free':
    f=f+1
  else:
    p=p+1
print('There are ',f, 'Free Applications',end=" ")
print('and',p, 'Paid Applications')
```

Find the Frequency of each Catagory in the table

```python
Category={}
for name in df['Category'].unique():
  Category[name]=0
for i in df['Category']:
  Category[i]+=1
Category
```
